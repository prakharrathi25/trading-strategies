## Feature Engineering

### Extract year

```python
data['sa_finance1_year']
```

```python
def extract_year(val):
    val = str(val)
    year = int(val[:4])
    return year
```

```python
data['sa_year'] = data['sa_finance1_year'].apply(extract_year)
```
